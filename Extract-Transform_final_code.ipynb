## Deliverable 1: Extract
----
### Option 2: Use regex 

1. Import the `backer_info.csv` file into a DataFrame. 
2. Extract the "backer_id", "cf_id", "name", and "email using regular expressions."
3. Create a new DataFrame with the retrieved data.
4. Export the DataFrame as `backers_data.csv`.

```python
# Get the backers_info from the crowdfunding_info sheet. 
pd.set_option('max_colwidth', 400)
backers = pd.read_csv("backer_info.csv")
backers
```

```python
# Extract the alpha-numeric "backer_id" from the backer_info column using a regex expression 
# and add it as a new column called "backer_id".
backers['backer_id'] = backers['backer_info'].str.extract('\"backer_id\"\:\s+\"(\w+)\"')

#print the df
backers

#test to make sure the backer_id extracted matches the backer_id put into the df.
    #a sum of 8175 (length of df) means it worked completely
#(backers['backer_info'].str.extract('\"backer_id\"\:\s+\"(\w+)\"')[0] == backers['backer_id']).sum()
```

```python
# Extract the two to four-digit "cf_id" number from the backer_info column. 
# and add it as a new column called "cf_id".
backers['cf_id'] = backers['backer_info'].str.extract('\"cf_id\"\:\s+(\d+)\W').astype(int)

#print the df
backers

#test to make sure the cf_id extracted matches the cf_id put into the df.
    #a sum of 8175 (length of df) means it worked completely
#(backers['backer_info'].str.extract('\"cf_id\"\:\s+(\d+)\W').astype(int)[0] == backers['cf_id']).sum()
```

```python
# Extract the name from the backer_info column and add it as a new column called "name".
backers['name'] = backers['backer_info'].str.extract('\"name\"\:\s+\"(\w+\s+\w+)\"')

#print the df
backers

#test to make sure the backer_id extracted matches the backer_id put into the df.
    #a sum of 8175 (length of df) means it worked completely
#(backers['backer_info'].str.extract('\"name\"\:\s+\"(\w+\s+\w+)\"')[0] == backers['name']).sum()
```

```python
# Extract the email from the backer_info column and add it as a new column called "email".
backers['email'] = backers['backer_info'].str.extract('\"(\S+@\S+)\"\W')
#use *.str.extractall(pat).values if you do extractall since it becomes multi-indexed and .values will make it a list

#print the df
backers

#test to make sure the email extracted matches the email put into the df.
    #a sum of 8175 (length of df) means it worked completely
#(backers['backer_info'].str.extract('\"(\S+@\S+)\"\W')[0] == backers['email']).sum()
```

```python
#check backers has the right datatypes
#backers.info()

# Create a new DataFrame with the appropriate columns.
backer_df_regex = backers[["backer_id", "cf_id", "name", "email"]].copy()
backer_df_regex
```

```python
# Export the DataFrame as a CSV file using encoding='utf8'.
backer_df_regex.to_csv("backer_df_regex.csv", index=False, encoding='utf-8')
```

```python
#we can check to make sure the two methods gave the same answers

#gives an array of true-false values
backer_df == backer_df_regex 

#check sums to quick-check each column (8175 means all values are true)
(backer_df == backer_df_regex).sum()
```
